Introdução aos Grafos - Python

```python
from igraph import Graph
from igraph import plot
```

```python
#Definição do grafo com as arestas
grafo1 = Graph(edges = [(0,1), (1,2), (2,3), (3,0)], directed = True)
#Definição do rótulo de cada vértice
grafo1.vs['label'] = range(grafo1.vcount())
print(grafo1)
```

```python
#gráfico
plot(grafo1, bbox=(0,0,300,300))
```

```python
#Criação do segundo grafo
grafo2 = Graph(edges = [(0,1), (1,2), (2,3), (3,0), (0,3), (3,2), (2,1), (1,0)], directed = True)
grafo2.vs['label'] = range(grafo2.vcount())
plot(grafo2, bbox=(0,0,400,400))
```

```python
#Grafo com Laço
grafo3 = Graph(edges = [(0,1),(1,2),(2,3),(3,0),(1,1)], directed = True)
grafo3.vs['label'] = range(grafo3.vcount())
plot(grafo3, bbox=(0,0,300,300))
```

```python
#Criação do quarto grafo
grafo4 = Graph(edges = [(0,1),(1,2),(2,3),(3,0),(1,1)], directed = True)
#adicionamos vertice isolado
grafo4.add_vertex(5)
grafo4.vs['label'] = range(grafo4.vcount())
plot(grafo4, bbox=(0,0,300,300))
```

```python
#grafo direcionado
grafo5 = Graph(edges = [(0,1),(1,2),(2,3),(3,0)], directed = True)
grafo5.vs['label'] = range(grafo5.vcount())
print(grafo5)
```

```python
#Criação não direcionado
grafo6 = Graph(edges = [(0,1),(1,2),(2,3),(3,0)], directed = False)
grafo6.vs['label'] = range(grafo6.vcount())
print(grafo6)
```

```python
#gráfico
plot(grafo6, bbox=(0,0,300,300))
```

```python
#Adicionando vértices e arestas por meio das funções add_vertices e add_vertex
grafo7 = Graph(directed = False)
grafo7.add_vertices(10)
#grafo7.add_vertex(16)
grafo7.add_edges([(0,1),(2,2),(2,3),(3,0)])
grafo7.vs['label'] = range(grafo7.vcount())
print(grafo7)
plot(grafo7, bbox=(0,0,300,300))
```

```python
#Criação do quarto grafo, configurando rótulos personalizados para os vértices
grafo8 = Graph(directed = False)
grafo8.add_vertices(5)
grafo8.add_edges([(0,1),(1,2),(2,3),(3,4),(4,0),(0,2),(2,1)])
grafo8.add_vertex(5)
grafo8.add_vertex(6)
grafo8.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
plot(grafo8, bbox=(0,0,300,300))
```

```python
grafo9 = Graph(directed = False)
grafo9.add_vertices(5)
grafo9.add_edges([(0,1),(1,2),(2,3),(3,4),(4,0),(0,2),(2,1)])
grafo9.add_vertex(5)
grafo9.add_vertex(6)
grafo9.vs['label'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
grafo9.vs['name'] = ['A', 'B', 'C', 'D', 'E', 'F', 'G']
```

```python
#Impressão da matriz de adjacência
print(grafo9.get_adjacency())
```

```python
#linha
grafo9.get_adjacency()[0,]
```

```python
#linha e coluna
grafo9.get_adjacency()[0,1]
```

```python
# Estrutura de repetição para percorrer cada vértice, visualizando o nome e o rótulo
for v in grafo9.vs:
   print(v) 
```

```python
plot(grafo9, bbox=(0,0,300,300))
```

```python
#Criação de grafo com pesos entre as relações
grafo10 = Graph(edges = [(0,1),(2,3),(0,2),(0,3)], directed = True)
grafo10.vs['label'] = ['Fernando', 'Pedro', 'Jose', 'Antonio']
grafo10.vs['peso'] = [40,30,30,25]
print(grafo10)
```

```python
#Percorrer os vértices para visualizar os pesos
for v in grafo10.vs:
    print(v)
```

```python
grafo10.vs[0]
```

```python
#Definição do tipo de amizade e do peso das relações
grafo10.es['TipoAmizade'] = ['Amigo', 'Inimigo', 'Inimigo', 'Amigo']
grafo10.es['weight'] = [1,2,1,3]
print(grafo10)
```

```python
#Percorrer os vértices, tipo de amizade
for e in grafo10.es:
    print(e)
```
